## 7.11 共享坐标轴

常常需要绘制包含相同坐标轴(x或y或都是)的两个或更多的图形。`plt.subplots`提供了一种简单的方法来绘制通用(共享的)的坐标轴。首先，我们将生成具有不同范围的虚拟数据。然后使用`plt.subplots`绘制图形。`plt.subplots`中的其他选项与`plt.subplot`相似。

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = range(100)
x2 = range(125)

y1 = np.random.rand(100)
y2 = 2.0*np.random.rand(125)
y3 = np.random.rand(125)
y4 = 1.5*np.random.rand(100)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True, sharey=True)
ax1.plot(x1,y1,'ro')
ax2.plot(x2,y2,'go')
ax3.plot(x2,y3,'bs')
ax4.plot(x1,y4,'mp')
plt.tight_layout()
plt.show()
```
