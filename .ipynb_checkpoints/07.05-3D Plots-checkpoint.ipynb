## 7.5 三维图

为了绘制三维图，我们需要从`mpl_toolkits.mplot3d`导入`Axes3D`库。散点图或折线图在三维中绘制的方法与二维类似。我们将生成三个变量，绘制三维散点图。图7.8显示三维散点图。

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='k', marker='s')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
plt.close()
```
