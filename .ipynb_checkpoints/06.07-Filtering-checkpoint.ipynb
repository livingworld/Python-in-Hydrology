## 6.7 过滤

有源雷达数据受到散斑/噪声的影响。在我们从卫星数据中提取有有用的信息之前，我们需要从数据中移除/最小化这些散斑。这些过滤器本质上是试图去除高频信息。实际上这种高频信息不必是噪音。所以我们需要指定想要多少过滤器。过滤器的类型可以划分为两类:自适应和非自适应过滤器。自适应过滤器将图像中的权重调整到散斑级别，而非自适应过滤器在整个影像上均匀地应用相同的权重。在本节，我们将这两个类别的一个例子。我们将使用非自适应过滤器类的中值过滤器，而使用自适应过滤器类中的Wiener过滤器。

我们可以从`scipy.signal`库中导入`metfilt2d`函数。首先，我们从硬盘中读取以tif格式存储的噪音数据。然后我们应用一个窗口大小为`3x3`的过滤器。图6.6显示了过滤后的影像。当我们将这个影像与原始影像比较，及与混杂了一些噪音的影像比较，我们可以看到过滤后的影像显示出更平滑的变化，但没有显示出接近原始图像的位置。

```python
from osgeo import gdal
from scipy.signal import medfilt2d
from osgeo.gdalconst import *
import matplotlib.pyplot as plt

# read the raster data
driver = gdal.GetDriverByName('GTiff')
file_name = "datas/data_noisy.tif"
dataset = gdal.Open(file_name, GA_ReadOnly)
geotransform = dataset.GetGeoTransform()
data = dataset.GetRasterBand(1).ReadAsArray()
dataset = None

data_median = medfilt2d(data, kernel_size=3) # median filter of 3X3 window
# plot the data
plt.matshow(data_median)
plt.colorbar()
plt.show()
```

现在，我们在相同的数据集上应用Wiener过滤器。我们也为Wiener过滤器保持相同的窗口大小。经过这样的过滤，我们也以tif格式保存数据。这说明，如何读取一些地理空间数据，处理特征信息(在这种情况下为矩阵)，然后用地理坐标信息保存特征信息。

```python
from scipy.signal import wiener
data_wiener = wiener(data, mysize=(3,3)) # Wiener filter
# plot
plt.matshow(data_wiener)
plt.colorbar()
plt.show()

# save the data into tif format
driver = gdal.GetDriverByName('GTiff')
file_name = "datas/data_filtered.tif"
dataset = driver.Create(file_name, data_wiener.shape[1], data_wiener.shape[0], 1, gdal.GDT_Float32)
dataset.SetGeoTransform(geotransform)
dataset.GetRasterBand(1).WriteArray(data_wiener, 0, 0)
dataset = None
```
