## 4.2 水蒸气

大致上，饱和蒸气压（$e_s$）与空气温度（$T$）通过以下等式相关。

<center>$e_s = 611exp(\frac{17.27T}{237.3+T})$, (4.1) </center>     

其中，$e_s$是帕斯卡,$T$为摄氏度。让我们计算$T＝50$时，$e_s$的值。

````python
import numpy as np

T = 50
es = 611*np.exp(17.27*T/(237.3+T))
print(es)
````

让我们在-100≤T≤100的范围内绘制$e_s$与$T$的变化曲线。`plt.plot(x，y)`绘制y与x的线图，默认颜色为蓝色。`plt.xlabel()`和`plt.ylabel()`分别用于x轴和y轴上写标签。`xlable`和`ylabel`的输入必须是一个字符串，或一个包含字符串的变量。`plt.show()`显示图形在电脑屏幕上。

````python
import numpy as np
import matplotlib.pyplot as plt
T = np.linspace(-100,100,50)
es = 611*np.exp(17.27*T/(237.3+T))
plt.plot(T,es)
plt.xlabel('T (degree Celcius)')
plt.ylabel('es (Pa)')
plt.show()
````

结果如上图所示。这个例子演示了如何用图形直观地显示一个变量相对于另一个变量的变化，而前者是后面的显式函数。

## 4.3 降雨

雨滴的最终速度($V_t$)由下式给出:

<center>$V_t = [\frac{4gD}{3C_d}(\frac{\rho_w}{\rho_a}-1)]^{1/2} $, (4-2)</center>

$g$是重力加速度，$D$是雨滴的直径，$\rho_w$是水的密度，$\rho_a$是空气密度，$C_d$是阻力系数。斯托克定律可以用来计算阻力系数（$C_d=24/R_e$）,这对于直径小于0.1mm的雨滴是有效的。$R_e$是雷诺数，能够以$\rho_aVD$的形式计算。我们假设，$R_e$是5.0，雨滴直径为0.05mm，我们想估计$V_t$。($\rho_w=998,\rho_a=1.2$)

````python
import numpy as np
Re = 5.0;rho_w = 998;rho_a = 1.2;g = 9.8;D = 0.05E-3
Cd = 24/Re
Vt = np.sqrt((4*g*D)/(3*Cd)*(rho_w/rho_a-1))
Vt
````

在这个例子中，我们看到“；”允许我们在一行中定义许多表达式。

通常，我们在相同的雨量计中得到雨量，我们要做雨量的等值线(轮廓线)图。为了说明这种情况，我们首先用随机数生成10个站点的位置(x,y)和雨量。生成的雨量计的位置如图所示。

````python
# 导入需要的模块
import numpy as np
import matplotlib.pyplot as plt

# 生成位置和雨量
x = np.random.rand(10)
y = np.random.rand(10)
rain = 10*np.random.rand(10)

#绘制位置
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('figures/loc.png')
plt.show()
````

我更喜欢在一段代码之后添加空白行，并在它的顶部做注释，这增加了代码的可读性。`plt.scatter()`作散点图，即绘制点而不是线。如果数据中的位置没有顺序，则使用散点图。像在这种情况下，可能有两个靠近的站，但可能被放置在远处的数组中。



绘制等高线的流程图如图4.5所示。首先，我们需要生成具有规则间距的网格，其大小与雨量计的位置相同。然后，从给定的位置和雨量数据中，我们需要使用一些插值方案在规则网格上计算数据。之后可以获得等值线图。`scipy.interpolate`库的`griddata`函数在获取网格数据(常规网格数据)时非常有用。当我们只需要库中的一个或几个函数时，最好明确地调用它们，例如`from scipy.interpolate import griddata`,就像下一个例子。我们使用`numpy`中的`meshgrid`函数，从给定的x和y向量中创建网格。

```flow
st=>start: Start
e=>end: End
op1=>operation:雨量计位置
op2=>operation:数据(例如雨量)
op3=>operation:插值
op1->op3
```

````python
from scipy.interpolate import griddata
#generate the desired grid, where rainfall is to be interpolated
X,Y = np.meshgrid(np.linspace(0,1,1000), np.linspace(0,1,1000))

#perform the gridding
grid_rain = griddata((x,y), rain, (X, Y))
````

现在，我们通过`plt.contourf()`函数制作网格数据的等高线图。`contourf`制作填充轮廓，而`contour()`提供简单的轮廓。尝试使用`contour`而不是`contourf`，你会看到其中的差别。我们使用`plt.clf()`来清除当前的图形，因为在内存中可能存在一些现有的图形，特别是如果您在同一会话中伴随所有的实例。我们也使用`plt.scatter()`来覆盖雨量计的位置。`s`和`c`分别用来定义标记的大小和颜色。`plt.xlim()`和`plt.ylim()`分别限制$x$轴和$y$轴的范围。

````python
plt.clf()
plt.contourf(X,Y,grid_rain)
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(x, y, s=30, c='r')
plt.xlim((0,1))
plt.ylim((0,1))
plt.savefig('figures/grid_rain.png')
plt.show()
````

## 4.5 蒸发

根据能量平衡，忽略感热通量和地热通量后，蒸发率($E_r$)可以计算为，

<center>$E_r=\frac{R_n}{l_v\rho_w}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.3)</center>

其中，$R_n$是净辐射，$l_v$是净汽化潜热，$\rho_w$是水的密度。$l_v$可以近似为，

<center>$l_v=2500-2.36×T$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.4)</center>

其中，$T$是摄氏度。

根据空气动力学，蒸发率（$E_a$）可以计算为，

<center>$E_a=B(e_{as}-e_a)$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.5)</center>

其中，

<center>$B=\frac{0.622k^2\rho_au_2}{p\rho_w[ln(z_2/z_0)]^2}$&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.6)</center>

$e_{as}$为饱和蒸气压，$e_a$是蒸气压，$k$是冯卡曼系数,$u_2$是在$z_2$米高度测量的风速，$p$是空气压力，$z_0$是粗糙度高度。

通常，通过结合能量平衡和空气动力学方法计算蒸发量。在这种情况下，$E$变为，

<center>$E=\frac{\Delta}{\Delta+\gamma}E_r+\frac{\gamma}{\Delta+\gamma}E_a$&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.7)</center>

其中，$\Delta$是饱和蒸汽压力曲线的梯度，

<center>$\Delta=\frac{4098e_s}{(273.3+T)^2}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.8)</center>

$\gamma$是湿度常数，被定义为，

<center>$\gamma=\frac{C_pK_hp}{0.622l_vK_w}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.9)</center>

$k_h$和$k_w$分别是热量和蒸汽扩散率。

让我们使用随机数生成模拟数据。我们知道`np.random.rand`在`[0，1]`区间内提供均匀分布的随机数。如果我们想在其他的范围内得到均匀分布的随机数，例如`[a,b]`，我们可以用下面的方式来转换变量:

<center>$x_{new}=a+(b-a)*x_{old}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.10)</center>

其中，$x_{old}$是`[0,1]`上的均匀随机变量，$x_{new}$有`[a,b]`的范围。`np.random.randn`提供均值为0，标准差为1的正态分布的随机变量。如果我们对平均值为$\mu$，标准差为$\gamma$的正态分布随机变量感兴趣，我们做下面的转换。

<center>$y_{new}=μ+\sigma*y_{old}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.11)</center>

其中，$y_{new}$是均值为$\mu$且标准差为$\sigma$的变换通量，$y_{old}$是由`np.random.randn`函数产生的具有均值为0且标准差为1的正态分布随机变量。

在下面的例子中，我们将生成通常范围内的变量。变量之后的注释提供了在均匀分布随机变量情况下的上下限，以及在变量为正态分布时的均值和标准差的细节。

````python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt

# 生成模拟数据
Rn = 150+100*np.random.rand(100) # 下限= 150, 上限 = 250
T = 25+3*np.random.randn(100) # 均值 = 25, 标准差 = 3
Rh = 0.2+0.6*np.random.rand(100) # 下限 = 0.2, 上限 = 0.8
u2 = 3+np.random.randn(100) # 均值 = 3, 标准差 = 1

# 定义常量
rho_w = 997; rho_a = 1.19; p = 101.1e3; z2 = 2
z0 = 0.03e-2; k = 0.4; Cp = 1005
````

现在，我们采用基于能量平衡的方法来估算蒸发量。

````python
lv = (2500-2.36*T)*1000 # 乘以1000将KJ/kg转换为J/kg
Er = 200/(lv*997)
Er *= 1000*86400 # 将 m/s 转换为 mm/day
````

我们使用乘法和赋值运算符来转换单位。我们也可以通过简单的乘法来完成，例如$E_r = E_r*1000*86400$。乘法和赋值运算速度很快，因为它不会在内存中创建任何临时变量。事实上，所有的赋值操作符都比简单操作符快，只要有作用域就可以使用。现在我们用空气动力学方法估算蒸发量。

````python
B = 0.622*k**2*rho_a*u2/(p*rho_w*(np.log(z2/z0))**2)
e_s = 611*np.exp(17.27*T/(237.3+T))
e_a = Rh*e_s
Ea = B*(e_s-e_a)
Ea *= 1000*86400 # 将 m/s 转换为 mm/day
````

将能量平衡和空气动力学方法相结合，对蒸发量进行改进。

````python
gamma = Cp*p/(0.622*lv) # 由于 kh/kw = 1, 因而他们被抵消了
delta = 4098*e_s/(237.3+T)**2
w = delta/(delta+gamma)
E = w*Er + (1-w)*Ea
````

现在，我们得到了四个重要的变量:能量平衡法得到蒸发量($E_r$)、空气动力学方法得到的蒸发量($E_a$)、组合蒸发量($E$)、能量平衡法蒸发比法($E_r/E$)。我们可以将这四个变量绘制在四个不同的图中，也可以将图形分成四个部分。`subplot`是使图分段的一种函数。`subplot`的第一个参数是所需的行数，第二个参数是理想的图形列数，第三个参数是`subplot`中的位置，这是从左到右，从上到下。

````python
plt.clf()
plt.subplot(2,2,1)
plt.plot(Er)
plt.xlabel('Time')
plt.ylabel('Er')

plt.subplot(2,2,2)
plt.plot(Ea)
plt.xlabel('Time')
plt.ylabel('Ea')

plt.subplot(2,2,3, facecolor ='y')
plt.plot(E)
plt.xlabel('Time')
plt.ylabel('E')

plt.subplot(2,2,4, facecolor ='g')
plt.plot(w)
plt.xlabel('Time')
plt.ylabel('Er/E')
plt.savefig('figures/E.png')
plt.show()
````

<center>图4.7:$E_r,E_a,E$和$E_r/E$的估值</center>

$E_r,E_a,E$和$E_r/E$的估值如图4.7所示。我们还是用`facecolor`定义`subplot`的背景颜色。

在图4.7中，子图4的`ylabel`与子图3重叠。这可以通过改变`wspace`校正。下面将对此进行说明。图4.8给出了改进的情节。

````python
plt.clf()
fig = plt.figure()
fig.subplots_adjust(wspace=0.6)
plt.subplot(2,2,1)
plt.plot(Er)
plt.xlabel('Time')
plt.ylabel('Er')

plt.subplot(2,2,2)
plt.plot(Ea)
plt.xlabel('Time')
plt.ylabel('Ea')

plt.subplot(2,2,3, facecolor='y')
plt.plot(E)
plt.xlabel('Time')
plt.ylabel('E')

plt.subplot(2,2,4, facecolor='g')
plt.plot(w)
plt.xlabel('Time')
plt.ylabel('Er/E')
plt.savefig('figures/corr_E.png')
plt.show()
````

<center>图4.8:校正ylabel$E_r,E_a,E$和$E_r/E$的估值</center>

## 4.6 浸润

Green-Ampt法给出的累积渗透量写成：

<center>$F(t)-\psi\Delta\theta{ln(1+\frac{F(t)}{\psi\Delta\theta})}=Kt$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.12)</center>

其中，$F(t)$是t时刻后的累积渗透量，$\psi$是吸头，$\Delta\theta$给定为，

<center>$\Delta\theta=(1-S_e)\theta_e$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.13)</center>

其中，$S_e$是饱和度，$\theta_e$是有效孔隙度，$K$是渗透系数。为了用迭代过程求解方程，式4.12被改写为，

<center>$F(t)=\psi\Delta\theta{ln(1+\frac{F(t)}{\psi\Delta\theta})}+Kt$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.14)</center>

我们使用`while`函数进行迭代，直达达到要求的精度。$F$的迭代值使用`append`方法存储。`append`逐个追加数组，并将输入变量放入其中。

````python
from __future__ import division
import numpy as np
# 定义变量
theta_e = 0.486
psi = 16.7
K = 0.65
S_e = 0.3
t = 1

# 计算dtheta
dtheta = (1-S_e)*theta_e

# F的初始估计
F_old = K*t
epsilon = 1
F = []
while epsilon > 1e-4:
    F_new = psi*dtheta * np.log(1+F_old/(psi*dtheta)) + K*t
    epsilon = F_new - F_old
    F_old = F_new
    F.append(F_new)
````

现在，我们绘制$F$迭代值的图，看看$F$是如何用迭代更新。我们也在`plot`函数中使用了`-ok`。`-o`表示填充点的连续线，`k`表示图的颜色为黑色。我们也指定`xlabel`和`ylabel`的字体大小。我们用给`ylabel`指定了‘25’，给`xlabel`指定了‘20’，只是表示不同大小的字体可用于不同的文本。当然，没有必要为`xlabel`和`ylabel`定义不同的字体大小。同样的参数`fontsize`可以用来定义图例的字体大小。

````python
import matplotlib.pyplot as plt
plt.plot(F,'-ok')
plt.xlabel('Number of iteration',fontsize=25)
plt.ylabel('F',fontsize=20)
plt.savefig('figures/F.png')
plt.show()
````

<center>图4.9:$F$随迭代次数的变化。</center>

## 4.7 地表水

矩形渠道中的流量深度由下式给出，

<center>$Q=\frac{1.49}{n}S_0^{1/2}\frac{(By)^{5/3}}{(B+y)^{2/3}}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.15)</center>

其中，$Q$为流量,$n$为曼宁系数,$S_0$为水面坡度，$B$是渠道宽度，$y$为流量深度。这是$y$中的一个非线性方程，还没有得到显式解。可以像上一节那样迭代解决，或者使用像牛顿迭代法(Newton-Raphson)这样的方法。在这里，我们将使用`Scipy.optimize`库中的`fmin`函数来解决。首先，我们将导入所需的库。然后，我们将定义一个函数，以流量深度(y)作为输入，返回基于这个y和给定的Q估计的流量的误差。我们取误差的绝对值，也可取其他如误差平方等。在指定函数之后，我们可以把这个函数和`y`的一些初始估计值作为`fmin`的输入。

````python
# 导入依赖库
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin

# 定义变量
n = 0.015
S0 = 0.025
Q = 9.26
B = 2

# 定义流量函数
def flow(y):
    Q_estiamted = (1.49/n)*(S0**0.5)*((B*y)**(5/3))/((B+y)**(2/3))
    epsilon = np.abs(Q_estiamted - Q)
    return epsilon

y_optimum = fmin(flow,0.5)
````

`fmin`给我们所需的y值。我们也可以得到迭代的详细信息，以及最终迭代的误差值。我们使用`print`打印y优化值如下。

````python
print(y_optimum)
````

优化成功终止，即在默认最大允许迭代次数内实现所需的精度。输出表明需要13次迭代才能达到所需的精度，并且在这个过程中对函数进行了26次求值。

## 河道马斯京根演算法

使用马斯京根法计算出流量如下，

<center>$Q_{j+1}=C_1I_{j+1}+C_2I_j+C_3Q_j$.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.16)</center>

我们给出了$C_1,C_2,C_3$和$Q_0$的值，并且我们有兴趣从时间(t=0)到(t=19)获得Q的值。在这个例子中，首选我们定义变量，然后使用`for`循环迭代。列表`I`很长，不适合一行。在这种情况下，我们也可以到第二行，开头括号告诉Python列表已经开始，而结尾括号告诉Python这是列表的结尾。我们可以根据需要在任意多行中编写列表，不需要指定任何其他内容来告诉Python列表是多行定义的。

````python
from __future__ import division
import numpy as np
# 定义变量
I = np.array([93, 137, 208, 320, 442, 546, 630, 678, 691, 675, 634, 571, 477,
390, 329, 247, 184, 134, 108, 90])
C1 = 0.0631
C2 = 0.3442
C3 = 0.5927

Q = np.empty(20) # 定义空数组
Q[0] = 85 # Q的初始值

# 循环
for i in range(1,20):
    Q[i] = C1*I[i] + C2*I[i-1] + C3*Q[i-1]
````

现在我们可以使用`matplotlib.pyplot`来绘制入流量和出流量。`-*`表示星形连续线，`--s`表示表示正方形虚线。

````python
import matplotlib.pyplot as plt
plt.plot(I, '-*', label='Inflow')
plt.plot(Q, '--s', label='Outflow')
plt.xlabel('Time', fontsize=20)
plt.ylabel('Flow', fontsize=20)
plt.legend()
plt.savefig('figures/muskingum.png')
plt.show()
````

<center>图4.10:入流量和出流量随时间的变化。</center>
