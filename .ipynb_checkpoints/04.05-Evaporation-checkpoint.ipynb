## 4.5 蒸发

根据能量平衡，忽略感热通量和地热通量后，蒸发率($E_r$)可以计算为，

<center>$E_r=\frac{R_n}{l_v\rho_w}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.3)</center>

其中，$R_n$是净辐射，$l_v$是净汽化潜热，$\rho_w$是水的密度。$l_v$可以近似为，

<center>$l_v=2500-2.36×T$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.4)</center>

其中，$T$是摄氏度。

根据空气动力学，蒸发率（$E_a$）可以计算为，

<center>$E_a=B(e_{as}-e_a)$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.5)</center>

其中，

<center>$B=\frac{0.622k^2\rho_au_2}{p\rho_w[ln(z_2/z_0)]^2}$&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.6)</center>

$e_{as}$为饱和蒸气压，$e_a$是蒸气压，$k$是冯卡曼系数,$u_2$是在$z_2$米高度测量的风速，$p$是空气压力，$z_0$是粗糙度高度。

通常，通过结合能量平衡和空气动力学方法计算蒸发量。在这种情况下，$E$变为，

<center>$E=\frac{\Delta}{\Delta+\gamma}E_r+\frac{\gamma}{\Delta+\gamma}E_a$&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.7)</center>

其中，$\Delta$是饱和蒸汽压力曲线的梯度，

<center>$\Delta=\frac{4098e_s}{(273.3+T)^2}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.8)</center>

$\gamma$是湿度常数，被定义为，

<center>$\gamma=\frac{C_pK_hp}{0.622l_vK_w}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.9)</center>

$k_h$和$k_w$分别是热量和蒸汽扩散率。

让我们使用随机数生成模拟数据。我们知道`np.random.rand`在`[0，1]`区间内提供均匀分布的随机数。如果我们想在其他的范围内得到均匀分布的随机数，例如`[a,b]`，我们可以用下面的方式来转换变量:

<center>$x_{new}=a+(b-a)*x_{old}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.10)</center>

其中，$x_{old}$是`[0,1]`上的均匀随机变量，$x_{new}$有`[a,b]`的范围。`np.random.randn`提供均值为0，标准差为1的正态分布的随机变量。如果我们对平均值为$\mu$，标准差为$\gamma$的正态分布随机变量感兴趣，我们做下面的转换。

<center>$y_{new}=μ+\sigma*y_{old}$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.11)</center>

其中，$y_{new}$是均值为$\mu$且标准差为$\sigma$的变换通量，$y_{old}$是由`np.random.randn`函数产生的具有均值为0且标准差为1的正态分布随机变量。

在下面的例子中，我们将生成通常范围内的变量。变量之后的注释提供了在均匀分布随机变量情况下的上下限，以及在变量为正态分布时的均值和标准差的细节。

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt

# 生成模拟数据
Rn = 150+100*np.random.rand(100) # 下限= 150, 上限 = 250
T = 25+3*np.random.randn(100) # 均值 = 25, 标准差 = 3
Rh = 0.2+0.6*np.random.rand(100) # 下限 = 0.2, 上限 = 0.8
u2 = 3+np.random.randn(100) # 均值 = 3, 标准差 = 1

# 定义常量
rho_w = 997; rho_a = 1.19; p = 101.1e3; z2 = 2
z0 = 0.03e-2; k = 0.4; Cp = 1005
```

现在，我们采用基于能量平衡的方法来估算蒸发量。

```python
lv = (2500-2.36*T)*1000 # 乘以1000将KJ/kg转换为J/kg
Er = 200/(lv*997)
Er *= 1000*86400 # 将 m/s 转换为 mm/day
```

我们使用乘法和赋值运算符来转换单位。我们也可以通过简单的乘法来完成，即$E_r = E_r*1000*86400$。乘法和赋值运算速度很快，因为它不会在内存中创建任何临时变量。事实上，所有的赋值操作符都比简单操作符快，只要有作用域就可以使用。现在我们用空气动力学方法估算蒸发量。

```python
B = 0.622*k**2*rho_a*u2/(p*rho_w*(np.log(z2/z0))**2)
e_s = 611*np.exp(17.27*T/(237.3+T))
e_a = Rh*e_s
Ea = B*(e_s-e_a)
Ea *= 1000*86400 # 将 m/s 转换为 mm/day
```

将能量平衡和空气动力学方法相结合，对蒸发量进行改进。

```python
gamma = Cp*p/(0.622*lv) # 由于 kh/kw = 1, 因而他们被抵消了
delta = 4098*e_s/(237.3+T)**2
w = delta/(delta+gamma)
E = w*Er + (1-w)*Ea
```

现在，我们得到了四个重要的变量:能量平衡法得到蒸发量($E_r$)、空气动力学方法得到的蒸发量($E_a$)、组合蒸发量($E$)、能量平衡法蒸发比法($E_r/E$)。我们可以将这四个变量绘制在四个不同的图中，也可以将图形分成四个部分。`subplot`是使图分段的一种函数。`subplot`的第一个参数是所需的行数，第二个参数是理想的图形列数，第三个参数是`subplot`中的位置，这是从左到右，从上到下。

```python
plt.clf()
plt.subplot(2,2,1)
plt.plot(Er)
plt.xlabel('Time')
plt.ylabel('Er')

plt.subplot(2,2,2)
plt.plot(Ea)
plt.xlabel('Time')
plt.ylabel('Ea')

plt.subplot(2,2,3, facecolor ='y')
plt.plot(E)
plt.xlabel('Time')
plt.ylabel('E')

plt.subplot(2,2,4, facecolor ='g')
plt.plot(w)
plt.xlabel('Time')
plt.ylabel('Er/E')
plt.savefig('figures/E.png')
plt.show()
```

<center>图4.7:$E_r,E_a,E$和$E_r/E$的估值</center>

$E_r,E_a,E$和$E_r/E$的估值如图4.7所示。我们还是用`facecolor`定义`subplot`的背景颜色。

在图4.7中，子图4的`ylabel`与子图3重叠。这可以通过改变`wspace`校正。下面将对此进行说明。图4.8给出了改进的情节。

```python
plt.clf()
fig = plt.figure()
fig.subplots_adjust(wspace=0.6)
plt.subplot(2,2,1)
plt.plot(Er)
plt.xlabel('Time')
plt.ylabel('Er')

plt.subplot(2,2,2)
plt.plot(Ea)
plt.xlabel('Time')
plt.ylabel('Ea')

plt.subplot(2,2,3, facecolor='y')
plt.plot(E)
plt.xlabel('Time')
plt.ylabel('E')

plt.subplot(2,2,4, facecolor='g')
plt.plot(w)
plt.xlabel('Time')
plt.ylabel('Er/E')
plt.savefig('figures/corr_E.png')
plt.show()
```
