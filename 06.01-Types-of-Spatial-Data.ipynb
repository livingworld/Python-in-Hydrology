## 6.1 空间数据类型

栅格和矢量是两种在GIS(地理信息系统)中存储和操作影像和图形数据的基础数据结构。栅格影像由单个的像素点形成，而且每个空间位置或分辨率元素拥有一个像素点。这个像素点关联的像素值表示属性，例如颜色、高程或ID值。矢量数据由点和线组成，这是几何与数学上的关联。使用坐标存储点，例如，二维点存储为(x,y)。线存储为一些列的点对，其中每一对代表一条直线段，例如(x1,y1)和(x2,y2)表示从(x1,y1)到(x2,y2)的线。

我们将使用一些数据函数创建一些栅格数据，然后也为它添加噪音。我们将保留所有数据(包含噪音和没有噪音的)到未来使用。`np.mgrid`被用于创建网格点。使用`plt.matshow`函数绘制数据，这个函数是一个简单的用来可视化二维数组的函数。没有噪音的数据如图6.1所示，被噪音污染的数据如图6.2所示。没有噪音的数据显示出系统的行为，而在添加噪音的数据中是模糊的。

```python
import numpy as np
import matplotlib.pyplot as plt 
# generate some synthetic data
X, Y = np.mgrid[0:101, 0:101]
data = np.sin((X**2 + Y**2)/25)
data_noisy = data + np.random.random(X.shape)

# plot the data
plt.matshow(data)
plt.colorbar()
plt.show()
```

<center>图6.1:虚拟数据创建</center>

```python
plt.matshow(data_noisy)
plt.colorbar(shrink=0.5)
plt.show()
```

<center>图6.2:噪声干扰下的虚拟数据</center>

我们也可以使用一些点来生成矢量数据。矢量数据如图6.3所示。

```python
# vector data
vector_x = [10,7,24,16,15,10]
vector_y = [10,23,20,14,7,10]

#plot vector data
plt.clf()
plt.plot(vector_x, vector_y)
plt.axis((5,25,5,25))
plt.show()
```
