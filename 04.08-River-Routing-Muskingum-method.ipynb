## 4.8 河道马斯京根演算法

使用马斯京根法计算出流量如下，

<center>$Q_{j+1}=C_1I_{j+1}+C_2I_j+C_3Q_j$.&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(4.16)</center>

我们给出了$C_1,C_2,C_3$和$Q_0$的值，并且我们有兴趣从时间(t=0)到(t=19)获得Q的值。在这个例子中，首选我们定义变量，然后使用`for`循环迭代。列表`I`很长，不适合一行。在这种情况下，我们也可以到第二行，开头括号告诉Python列表已经开始，而结尾括号告诉Python这是列表的结尾。我们可以根据需要在任意多行中编写列表，不需要指定任何其他内容来告诉Python列表是多行定义的。

```python
from __future__ import division
import numpy as np
# 定义变量
I = np.array([93, 137, 208, 320, 442, 546, 630, 678, 691, 675, 634, 571, 477,
390, 329, 247, 184, 134, 108, 90])
C1 = 0.0631
C2 = 0.3442
C3 = 0.5927

Q = np.empty(20) # 定义空数组
Q[0] = 85 # Q的初始值

# 循环
for i in range(1,20):
    Q[i] = C1*I[i] + C2*I[i-1] + C3*Q[i-1]
```

现在我们可以使用`matplotlib.pyplot`来绘制入流量和出流量。`-*`表示星形连续线，`--s`表示表示正方形虚线。

```python
import matplotlib.pyplot as plt
plt.plot(I, '-*', label='Inflow')
plt.plot(Q, '--s', label='Outflow')
plt.xlabel('Time', fontsize=20)
plt.ylabel('Flow', fontsize=20)
plt.legend()
plt.savefig('figures/muskingum.png')
plt.show()
```
